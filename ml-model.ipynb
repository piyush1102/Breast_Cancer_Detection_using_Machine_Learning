Dropping Target column

```python
x = df.drop(['target'], axis = 1)
x.head(6)
```

```python
y=df.target
y.head()
```

Splting Data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x1_sc=sc.fit_transform(x_train)
x2_sc=sc.transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=10000)
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
accuracy_score(y_test,pred)
```

```python
lr1 = LogisticRegression()
lr1.fit(x1_sc, y_train)
pred1 = lr1.predict(x2_sc)
accuracy_score(y_test, pred1)
```

```python
 from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,metric = 'minkowski',p=2)
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)
accuracy_score(y_test,knn_pred)
```

```python
knn1=KNeighborsClassifier(n_neighbors=5,metric = 'minkowski',p=2)
knn1.fit(x1_sc,y_train)
pred_knn1=knn1.predict(x2_sc)
accuracy_score(y_test,pred_knn1)
```

```python
from sklearn.svm import SVC
svm=SVC() 
svm.fit(x_train,y_train)
pred_svm=svm.predict(x_test)
accuracy_score(y_test,pred_svm)
```

```python
svm1=SVC()
svm1.fit(x1_sc,y_train)
pred_svm1=svm1.predict(x2_sc)
accuracy_score(y_test,pred_svm1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf.fit(x_train,y_train)
pred_rf=rf.predict(x_test)
accuracy_score(y_test,pred_rf)
```

```python
rf1=RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf1.fit(x1_sc,y_train)
pred_rf1=rf1.predict(x2_sc)
accuracy_score(y_test,pred_rf1)
```

```python
cm = confusion_matrix(y_test, pred_rf1)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
```

```python
print(classification_report(y_test, pred_rf1))
```

```python
from sklearn.model_selection import cross_val_score
cross_validation = cross_val_score(estimator = rf1, X = x1_sc, y = y_train, cv = 10)
print("Cross validation of Random Forest model = ",cross_validation)
print("Cross validation of Random Forest (in mean) = ",cross_validation.mean())
```

```python
from sklearn.model_selection import cross_val_score
cross_validation = cross_val_score(estimator = rf, X = x1_sc,y = y_train, cv = 10)
print("Cross validation accuracy of Random Forest model = ", cross_validation)
print("\nCross validation mean accuracy of Random Forest model = ", cross_validation.mean())
```
